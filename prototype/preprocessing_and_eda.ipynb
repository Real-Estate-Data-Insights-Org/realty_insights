## Exploratory Data Analysis (EDA)

We start by understanding the distribution of the target variable **SalePrice**.

```python
# Distribution of SalePrice

import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.hist(df['SalePrice'], bins=50)
plt.title("Sale Price Distribution")
plt.xlabel("SalePrice")
plt.ylabel("Number of Houses")
plt.show()
```

The plot shows strong right skew (few very expensive houses).
That’s why the dataset already contains LogSalePrice.

```python
# Distribution of LogSalePrice

plt.figure(figsize=(6,4))
plt.hist(df['LogSalePrice'], bins=50)
plt.title("Log Sale Price Distribution")
plt.xlabel("LogSalePrice")
plt.ylabel("Number of Houses")
plt.show()
```

This looks much more normal.

### Relationship Between House Size and Sale Price

Larger houses usually sell for higher prices.  
We check the relationship between **TotalFinishedArea** and **SalePrice**.

```python
# Relationship between TotalFinishedArea and SalePrice

plt.figure(figsize=(6,4))
plt.scatter(df['TotalFinishedArea'], df['SalePrice'], alpha=0.5)
plt.title("House Size vs Sale Price")
plt.xlabel("Total Finished Area")
plt.ylabel("SalePrice")
plt.show()
```

You can clearly see outliers (very expensive houses).

### Correlation Between Numerical Features

Correlation helps us understand how strongly variables are related to each other.

We especially check which features influence **SalePrice** the most.

```python
# Correlation matrix for numerical variables

corr_matrix = df.corr(numeric_only=True)

corr_matrix
```

Now we make it much easier to understand using a heatmap.

### Correlation Heatmap

A heatmap visually shows the correlation between numerical features.

Darker colors indicate stronger relationships between variables.

```python
# Heatmap of correlations

import seaborn as sns

plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

### Insight

From the correlation heatmap we observe:

- **TotalFinishedArea** shows a positive relationship with **SalePrice**, meaning larger houses tend to have higher prices.
- **TotalAppraisedValue** also has a noticeable positive correlation with **SalePrice**.
- Most other variables show **weak correlations**, indicating they may have limited direct influence on the sale price.
- The variable **LogSalePrice** is strongly related to **SalePrice**, as expected since it is a log transformation of the same variable.

Overall, house **size and appraised value appear to be the most important predictors of price** among the numerical features.

### Features Most Correlated With SalePrice

```python
# Correlation with SalePrice

corr_matrix['SalePrice'].sort_values(ascending=False)
```

### Insight

- **LogSalePrice** has the strongest correlation with SalePrice.
- Some location/building variables show moderate influence.
- Many features have weak correlation with price.

### Property Type Distribution

```python
# Distribution of property types

df['AssrLandUse'].value_counts().plot(kind='bar', figsize=(6,4))

plt.title("Property Type Distribution")
plt.xlabel("Land Use Type")
plt.ylabel("Count")
plt.show()
```

### Insight

- **ONE FAMILY** properties are the most common.
- **CONDOMINIUM** and **THREE FAMILY** are also frequent.
- Other property types appear much less often.

One-family homes dominate the dataset

### Sale Price by Property Type

```python
# Average sale price by property type

df.groupby('AssrLandUse')['SalePrice'].mean().sort_values().plot(kind='bar', figsize=(6,4))

plt.title("Average Sale Price by Property Type")
plt.xlabel("Property Type")
plt.ylabel("Average Sale Price")
plt.show()
```

### Insight

- **APT CRDA** and **CONDOMINIUM** properties have the highest average prices.
- **ONE FAMILY** and **TWO FAMILY** homes have moderate prices.
- **CONDO GARAGE** properties have the lowest average price.

### Land Size vs Sale Price

```python
# Relationship between land size and sale price

plt.figure(figsize=(6,4))
plt.scatter(df['LandSF'], df['SalePrice'], alpha=0.5)

plt.title("Land Size vs Sale Price")
plt.xlabel("Land Square Feet")
plt.ylabel("SalePrice")
plt.show()
```

### Insight

- There is **no strong clear relationship** between land size and sale price.
- Most houses cluster at **smaller land sizes with varied prices**.

## EDA Summary

### Average Sale Price by Neighborhood

### Insight

- This shows **location strongly influences property value**.

### Sale Price Distribution for Top 10 Neighborhoods

```python
# Top 10 neighborhoods by average price

top_neigh = df.groupby('xrPrimaryNeighborhoodID')['SalePrice'].mean().sort_values(ascending=False).head(10).index

plt.figure(figsize=(8,5))

sns.boxplot(
    x='xrPrimaryNeighborhoodID',
    y='SalePrice',
    data=df[df['xrPrimaryNeighborhoodID'].isin(top_neigh)]
)

plt.title("Sale Price Distribution for Top 10 Neighborhoods")
plt.xticks(rotation=45)
plt.show()
```

### Insight

- The top neighborhoods show **higher median sale prices**.
- Some neighborhoods also show **greater price variability**.
- Location clearly impacts property value.
