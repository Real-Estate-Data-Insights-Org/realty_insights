```python
import pandas as pd

df = pd.read_csv("preprocessing_eda_done_v3.csv")

df.head()
```

```python
# convert date column
df["SaleDate"] = pd.to_datetime(df["SaleDate"])

# check info
df.info()
```

```python
# create year and month from sale date
df["SaleYear"] = df["SaleDate"].dt.year
df["SaleMonth"] = df["SaleDate"].dt.month

# price per sqft
df["PricePerSqft"] = df["SalePrice"] / df["TotalFinishedArea"]

# appraisal ratio
df["AppraisalRatio"] = df["SalePrice"] / df["TotalAppraisedValue"]

df.head()
```

```python
# Remove identifier and text columns that do not contribute to analytical modeling
df = df.drop(columns=[
    "OBJECTID",
    "PropertyID",
    "ParcelID",
    "OwnerName",
    "PrimaryGrantor",
    "LegalReference",
    "LocationStreetName"
], errors="ignore")

df.head()
```

```python
# Convert categorical land use type into numeric codes
df["AssrLandUse"] = df["AssrLandUse"].astype("category").cat.codes

df.head()
```

```python
# Compute correlation between features and target variable
corr = df.corr(numeric_only=True)

corr["LogSalePrice"].sort_values(ascending=False)
```

```python
# Select features strongly correlated with LogSalePrice
important = corr["LogSalePrice"].abs() > 0.2

selected_features = corr.columns[important]

df_selected = df[selected_features]

df_selected.head()
```

```python
# Display final selected features for modeling
print("Selected Features:")
print(selected_features)
```
