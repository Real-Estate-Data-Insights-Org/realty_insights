```python
import pandas as pd

# Load dataset
df = pd.read_csv("feature_engineered_dataset.csv")

# Shape
print("Dataset Shape:", df.shape)

# First rows
df.head()
```

```python
# Feature and Target Selection

# Target variable
y = df['SalePrice']

# Drop non-predictive or identifier columns
X = df.drop(columns=[
    'SalePrice',
    'OwnerLastName',
    'StreetNameAndWay',
    'SaleDate'
], errors='ignore')

print("Feature Matrix Shape:", X.shape)
print("Target Shape:", y.shape)

print("\nFeatures Used for Model:")
print(X.columns)
```

### Insight

The target variable selected for prediction is **SalePrice**, which represents the final selling price of the property. Certain variables such as LogSalePrice, PricePerSqft, and AppraisalRatio were removed because they are derived from the target variable and could lead to data leakage.

```python
# Remove features that leak information about the target

leakage_features = ['LogSalePrice', 'PricePerSqft', 'AppraisalRatio']

X = X.drop(columns=[col for col in leakage_features if col in X.columns])

print("Updated Feature Matrix Shape:", X.shape)

print("\nFinal Features Used:")
print(X.columns)
```

```python
# Encode categorical variables and split dataset

from sklearn.model_selection import train_test_split

# Convert categorical columns to numerical using One-Hot Encoding
X_encoded = pd.get_dummies(X, drop_first=True)

print("Encoded Feature Shape:", X_encoded.shape)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded,
    y,
    test_size=0.2,
    random_state=42
)

print("\nTraining set shape:", X_train.shape)
print("Testing set shape:", X_test.shape)
```

### Insight

The dataset was divided into training and testing sets using an 80-20 split. The training data is used to train the machine learning models, while the testing data is used to evaluate how well the models generalize to unseen data.

```python
# Model Selection and Training

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

# Initialize models
lr = LinearRegression()
dt = DecisionTreeRegressor(random_state=42)
rf = RandomForestRegressor(random_state=42)

# Train models
lr.fit(X_train, y_train)
dt.fit(X_train, y_train)
rf.fit(X_train, y_train)

print("All models trained successfully!")
```

### Insight

Three regression models were trained: Linear Regression, Decision Tree, and Random Forest. These models were selected to compare simple linear relationships with more complex nonlinear models.

```python
# Model Prediction

lr_pred = lr.predict(X_test)
dt_pred = dt.predict(X_test)
rf_pred = rf.predict(X_test)

print("Predictions completed!")

print("\nSample Predictions (Random Forest):")
print(rf_pred[:10])
```

```python
# Model Evaluation

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
import pandas as pd

def evaluate_model(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    return mae, rmse, r2

# Evaluate models
lr_mae, lr_rmse, lr_r2 = evaluate_model(y_test, lr_pred)
dt_mae, dt_rmse, dt_r2 = evaluate_model(y_test, dt_pred)
rf_mae, rf_rmse, rf_r2 = evaluate_model(y_test, rf_pred)

# Create comparison table
results = pd.DataFrame({
    "Model": ["Linear Regression", "Decision Tree", "Random Forest"],
    "MAE": [lr_mae, dt_mae, rf_mae],
    "RMSE": [lr_rmse, dt_rmse, rf_rmse],
    "R2 Score": [lr_r2, dt_r2, rf_r2]
})

results
```

### Insight

The evaluation results show that Random Forest significantly outperforms the other models with the highest R² score and lowest prediction error. This indicates that ensemble tree-based models are better suited for capturing complex patterns in real estate pricing data.

```python
# Model Comparison Visualization

import matplotlib.pyplot as plt

plt.figure()

models = results['Model']
scores = results['R2 Score']

plt.bar(models, scores)

plt.title("Model Comparison (R2 Score)")
plt.xlabel("Models")
plt.ylabel("R2 Score")

plt.show()
```

```python
# Feature Importance from the best model (Random Forest)

import pandas as pd
import matplotlib.pyplot as plt

feature_importance = pd.Series(rf.feature_importances_, index=X_encoded.columns)

feature_importance = feature_importance.sort_values(ascending=False)

print("Top Important Features:")
print(feature_importance.head(10))

plt.figure()
feature_importance.head(10).plot(kind='bar')

plt.title("Top 10 Important Features Affecting Sale Price")
plt.xlabel("Features")
plt.ylabel("Importance Score")

plt.show()
```

### Insight

Feature importance analysis shows that **TotalAppraisedValue** is the most influential factor affecting property sale price. Location-related features such as neighborhood and property characteristics like finished area also contribute to the model’s predictions.

## Conclusion / Interpretation of Results

### Model Performance
Three regression models were trained and evaluated to predict property sale prices:
- Linear Regression
- Decision Tree
- Random Forest

The models were evaluated using MAE, RMSE, and R² score.

Among these models, Random Forest performed the best with an R² score of approximately 0.98. This indicates that the model explains around 98% of the variance in property sale prices. Decision Tree also showed strong performance with an R² score around 0.89, while Linear Regression performed poorly due to the nonlinear relationships present in real estate data.

### Feature Importance Insights
Feature importance analysis from the Random Forest model revealed that TotalAppraisedValue is the most influential factor affecting property sale price. Other important features include LocationStartNumber, PrimaryNeighborhoodID, and TotalFinishedArea.

This shows that both property characteristics and location-based factors significantly influence housing prices.

### Business Insights
From a business analytics perspective, the results indicate that:
- Appraised value is a strong indicator of market price
- Location and neighborhood significantly influence property valuation
- Property size contributes to price but less than location and appraisal value

### Final Model Selection
Based on model comparison and evaluation metrics, Random Forest was selected as the final model for predicting property sale prices because it achieved the highest R² score and lowest prediction error.
