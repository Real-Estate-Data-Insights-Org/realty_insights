```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("feature_engineered_dataset.csv")

print("Dataset Shape:", df.shape)

df.head()
```

```python
# Feature and Target Selection

# Target variable
y = df['SalePrice']

# Remove target and unnecessary columns
X = df.drop(columns=[
    'SalePrice',
    'OwnerLastName',
    'StreetNameAndWay',
    'SaleDate',
    'LogSalePrice',
    'PricePerSqft',
    'AppraisalRatio'
], errors='ignore')

print("Feature Matrix Shape:", X.shape)
print("Target Shape:", y.shape)

print("\nFeatures used:")
print(X.columns)
```

```python
# Encode categorical variables using One-Hot Encoding

X_encoded = pd.get_dummies(X, drop_first=True)

print("Encoded Feature Shape:", X_encoded.shape)

X_encoded.head()
```

```python
# Train-Test Split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded,
    y,
    test_size=0.2,
    random_state=42
)

print("Training set shape:", X_train.shape)
print("Testing set shape:", X_test.shape)
```

```python
# KNN Model Training

from sklearn.neighbors import KNeighborsRegressor

# Initialize model
knn = KNeighborsRegressor(n_neighbors=5)

# Train model
knn.fit(X_train, y_train)

print("KNN model trained successfully!")
```

```python
# Predict using KNN model

knn_pred = knn.predict(X_test)

print("Predictions completed!")

print("\nFirst 10 Predictions:")
print(knn_pred[:10])
```

```python
# Evaluate KNN model

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

mae = mean_absolute_error(y_test, knn_pred)
rmse = np.sqrt(mean_squared_error(y_test, knn_pred))
r2 = r2_score(y_test, knn_pred)

print("KNN Model Performance")
print("MAE:", mae)
print("RMSE:", rmse)
print("R2 Score:", r2)
```

```python
# Actual vs Predicted Values Plot

import matplotlib.pyplot as plt

plt.figure()

plt.scatter(y_test, knn_pred)

plt.xlabel("Actual Sale Price")
plt.ylabel("Predicted Sale Price")
plt.title("Actual vs Predicted Sale Price (KNN Model)")

plt.show()
```

```python
# Residual Plot

residuals = y_test - knn_pred

plt.figure()

plt.scatter(knn_pred, residuals)

plt.axhline(y=0)

plt.xlabel("Predicted Sale Price")
plt.ylabel("Residuals (Actual - Predicted)")
plt.title("Residual Plot (KNN Model)")

plt.show()
```

### Insight

K-Nearest Neighbors (KNN) regression was applied as an additional machine learning method to predict property sale prices. 
The model achieved an R² score of approximately 0.89, indicating that it can explain about 89% of the variance in the dataset.

KNN predicts values based on similarity between properties, meaning houses with similar features tend to have similar prices. 
While the model performed reasonably well, its performance was slightly lower than ensemble models such as Random Forest, 
which are better at capturing complex nonlinear relationships in the housing data.
